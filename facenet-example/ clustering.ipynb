```python
from keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from keras.preprocessing import image
from keras.models import load_model
from keras.applications.resnet50 import ResNet50
from keras.applications.xception import Xception
from PIL import Image
from sklearn.cluster import KMeans
import pyclustering
from pyclustering.cluster import xmeans
import numpy as np
import cv2
import glob
from tqdm import tqdm
import os
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox 
%matplotlib inline
```

```python
#画像を取得していきます。1行目では、dataフォルダのjpg写真のパスを全て取得します。
image_path = glob.glob('./data/*.jpg')
print('画像の枚数:',len(image_path))
```

```python
#学習済みモデルをダウンロードします。VGGを利用する場合は、2行目のコメントアウトをしてください。
model = ResNet50(include_top=False, weights='imagenet')

#model = Xception(include_top = False, weights = "imagenet")
#model = VGG16(weights='imagenet', include_top=False)
```

```python
#画像を学習済みモデルに通して、特徴量を抽出します。
X = []
for i in tqdm(range(len(image_path))):
    img = image.load_img(image_path[i], target_size=(224, 224))  # resize
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)  
    x = preprocess_input(x)  
    feat = model.predict(x) 
    feat = feat.flatten()
    X.append(feat)
```

```python
#取得した特徴量を2次元に削減します。
from sklearn.decomposition import PCA
pca = PCA(n_components=2) #次元数を2にしています
pca.fit(X)
reduced = pca.fit_transform(X)
print(reduced.shape) 
```

```python
#クラスタリングをしていきます。
from sklearn.cluster import KMeans
K = 3 #分類数を指定します。
kmeans = KMeans(n_clusters=K).fit(reduced)
pred_label = kmeans.predict(reduced)
```

```python
# クラスタリングした結果をプロット
x = reduced[:, 0]
y = reduced[:, 1]
plt.scatter(x, y, c=pred_label)
plt.colorbar()
plt.show()

#画像で再度表示します。
def imscatter(x, y, image_path, ax=None, zoom=1): 
    if ax is None: 
        ax = plt.gca() 

    artists = [] 
    for x0, y0,image in zip(x, y,image_path): 
        image = plt.imread(image) 
        im = OffsetImage(image, zoom=zoom) 
        ab = AnnotationBbox(im, (x0, y0), xycoords='data', frameon=False) 
        artists.append(ax.add_artist(ab)) 
    return artists

x = reduced[:, 0]
y = reduced[:, 1]
fig, ax = plt.subplots() 
imscatter(x, y, image_path, ax=ax,  zoom=.2) 
ax.plot(x, y, 'ko',alpha=0) 
ax.autoscale() 
plt.show() 
```

```python
#自動で分類数を算出します。
init_center = pyclustering.cluster.xmeans.kmeans_plusplus_initializer(reduced, 2).initialize()
xm = pyclustering.cluster.xmeans.xmeans(reduced, init_center, ccore=False)
xm.process()
clusters = xm.get_clusters()
print('クラスタ数:',len(clusters))
pyclustering.utils.draw_clusters(reduced, clusters, display_result=False)

x = reduced[:, 0]
y = reduced[:, 1]
fig, ax = plt.subplots() 
imscatter(x, y, image_path, ax=ax,  zoom=.2) 
ax.plot(x, y, 'ko',alpha=0) 
ax.autoscale() 
plt.show() 
```
